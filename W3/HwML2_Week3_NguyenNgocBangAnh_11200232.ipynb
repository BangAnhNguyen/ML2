```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Implement k means by numpy

```python
def k_means(X, k, centers=None, num_iter=100):
    if centers is None:
        rnd_centers_idx = np.random.choice(np.arange(X.shape[0]), k, replace=False)
        centers = X[rnd_centers_idx]
    for _ in range(num_iter):
        distances = np.sqrt(np.sum((X-centers[:,np.newaxis])**2, axis=-1))
        cluster_assignments = np.argmin(distances, axis=0)
        for i in range(k):
            msk = (cluster_assignments == i)
            centers[i] = np.mean(X[msk], axis=0) if np.any(msk) else centers[i]

    return cluster_assignments, centers
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
```

#### iris plants dataset

```python
iris = datasets.load_iris()
```

```python
iris_reduced = PCA(n_components = 2).fit_transform(iris.data)
```

```python
kmean_iris = k_means(X = iris_reduced, k = 2)
plt.scatter(iris_reduced[:, 0], iris_reduced[:, 1])
plt.scatter(kmean_iris[1][:, 0], kmean_iris[1][:, 1], c='r', s = 100)
```

```python
kmean_iris = k_means(X = iris_reduced, k = 3)
plt.scatter(iris_reduced[:, 0], iris_reduced[:, 1])
plt.scatter(kmean_iris[1][:, 0], kmean_iris[1][:, 1], c='r', s = 100)
```

```python
kmean_iris = k_means(X = iris_reduced, k = 4)
plt.scatter(iris_reduced[:, 0], iris_reduced[:, 1])
plt.scatter(kmean_iris[1][:, 0], kmean_iris[1][:, 1], c='r', s = 100)
```
